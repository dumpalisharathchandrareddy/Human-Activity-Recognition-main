```python
#labelencoding
from sklearn.preprocessing import LabelEncoder
var_mod = ['tBodyAcc-mean()-X', 'tBodyAcc-mean()-Y', 'tBodyAcc-mean()-Z',
       'tGravityAcc-mean()-X', 'tGravityAcc-mean()-Y', 'tGravityAcc-mean()-Z',
       'tBodyAccJerk-mean()-X', 'tBodyAccJerk-mean()-Y', 'tBodyAccJerk-mean()-Z',
       'tBodyGyro-mean()-X', 'tBodyGyro-mean()-Y', 'tBodyGyro-mean()-Z',
       'fBodyAcc-mean()-X', 'fBodyAcc-mean()-Y', 'fBodyAcc-mean()-Z',
       'fBodyGyro-mean()-X', 'fBodyGyro-mean()-Y', 'fBodyGyro-mean()-Z', 'label']

le = LabelEncoder()
for i in var_mod:
    if i == 'label':
        data[i] = le.fit_transform(data[i]).astype(int)
```

```python
data
```

```python
data

# Extract only the columns in var_mod as features
X = data[var_mod].drop("label", axis=1)

# Label encoding for the label column
y = data["label"]
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
data.label.unique()
```

```python
#X = data.drop(labels='label', axis=1)
#Response variable
#y = data.loc[:,'label']
```

```python
#We'll use a test size of 30%. We also stratify the split on the response variable, which is very important to do because there are so few fraudulent transactions.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
print("Number of features in the training data:", X_train.shape[1])
```

```python
#Random Forest
```

```python
#import libraries

from sklearn.ensemble import RandomForestClassifier
```

```python
#Decision Tree Classifier Initialization
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="weighted")
recall = recall_score(y_test, y_pred, average="weighted")
f1 = f1_score(y_test, y_pred, average="weighted")

# Print the performance metrics
print("Accuracy: {:.2f}".format(accuracy))
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1 Score: {:.2f}".format(f1))
```

```python
#Classification Report
print('Classification report Random Forest classifier Results:')
print(classification_report(y_test,y_pred))
```

```python
#Accuracy Score
x = (accuracy_score(y_test,y_pred)*100)
print('Accuracy result of Random Forest is:', x)
```

```python
cm2=confusion_matrix(y_test,y_pred)
print('Confusion Matrix result of Random Forest Classifier is:\n',cm2)
```

```python
specificity_1 = cm2[0,0]/(cm2[0,0]+cm2[0,1])
print('Specificity : ', specificity_1)
```

```python
sensitivity_2 = cm2[1,1]/(cm2[1,0]+cm2[1,1])
print('Sensitivity : ', sensitivity_2)
```

```python
def calc_sensitivity(confusion_matrix):
    """Calculate sensitivity (true positive rate) from a confusion matrix."""
    tp = confusion_matrix[1, 1]
    fn = confusion_matrix[1, 0]
    sensitivity = tp / (tp + fn)
    return sensitivity

def calc_specificity(confusion_matrix):
    """Calculate specificity (true negative rate) from a confusion matrix."""
    tn = confusion_matrix[0, 0]
    fp = confusion_matrix[0, 1]
    specificity = tn / (tn + fp)
    return specificity
confusion_matrix=cm2
sensitivity = calc_sensitivity(confusion_matrix)
specificity = calc_specificity(confusion_matrix)
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
```

```python
# perform k-fold cross-validation with k=accuracy
accuracy = cross_val_score(rfc, X, y, scoring='accuracy')
print('Cross validation test results of accuracy:')
print(accuracy)
```

```python
# Perform k-fold cross-validation with k=5
scores = cross_val_score(rfc, X, y, cv=5)

# Print the mean and standard deviation of the scores
print("Mean score: {:.2f}".format(scores.mean()))
print("Standard deviation: {:.2f}".format(scores.std()))
print(scores)
```
